# Exploratory Data Analysis (EDA) on `promise_exp.csv`

This notebook performs basic Exploratory Data Analysis on the `promise_exp.csv` dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Load Data

```python
import os

DATA_URL = 'https://raw.githubusercontent.com/egbertbritannia-cpu/rbl-requirements-classification/main/data/exp/promise_exp.csv'
file_path = DATA_URL

df = pd.read_csv(file_path)
display(df.head())
print(f"Loaded PROMISE expanded dataset from: {file_path}")
print(f"Total rows: {len(df)}")
```

## 2. Basic Information & Missing Values

```python
df.info()
```

```python
print("Missing values:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

## 3. Class Distribution

```python
class_counts = df['class'].value_counts(dropna=False)
display(class_counts)

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='class', order=class_counts.index, palette='viridis')
plt.title('Distribution of Requirements Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## 4. Text Length Analysis

```python
df['text_length'] = df['RequirementText'].astype(str).apply(len)
df['word_count'] = df['RequirementText'].astype(str).apply(lambda x: len(x.split()))

display(df[['text_length', 'word_count']].describe())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

sns.histplot(df['text_length'], bins=50, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Requirement Text Length (Characters)')

sns.histplot(df['word_count'], bins=50, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribution of Requirement Word Count')

plt.tight_layout()
plt.show()
```

## 5. Projects Analysis

```python
if 'ProjectID' in df.columns:
    project_counts = df['ProjectID'].value_counts()
    print(f"Total unique projects: {len(project_counts)}")
    
    plt.figure(figsize=(14, 5))
    sns.countplot(data=df, x='ProjectID', order=project_counts.index, palette='mako')
    plt.title('Number of Requirements per Project')
    plt.xlabel('Project ID')
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```
